```python
import pandas as pd
from matplotlib import pyplot
```

## Network Earthquake

```python
proba_exacte0 = [0.7148927097115598, 0.28510729028844006]

df_lw0 = pd.read_csv("lw_Earthquake.csv")
df_gibbs0 = pd.read_csv("gibbs_Earthquake.csv")
somme_lw = 0
somme_g = 0
erreur_lw = list()
erreur_g = list()
for i in range (len(df_gibbs0)):
    somme_lw += df_lw0.iloc[i]["prob_false"]
    somme_g += df_gibbs0.iloc[i]["prob_false"]
    cumul_proba_lw = somme_lw / (i+1)
    cumul_proba_g = somme_g / (i+1)
    erreur_lw.append(abs(cumul_proba_lw - proba_exacte0[0]))
    erreur_g.append(abs(cumul_proba_g - proba_exacte0[0]))
tab_y = [(n+1)*1000 for n in range(1000)]
pyplot.plot(tab_y, erreur_lw, color = 'red', linestyle = 'dashed', linewidth = 2)
pyplot.plot(tab_y, erreur_g, color = 'blue', linestyle = 'dashed', linewidth = 2)
pyplot.axis([-1000, 999999, 0, 0.3])
pyplot.title("Network Earthquake - erreur")
```

```python
time_lw = pd.read_csv("lw_time_Earthquake.csv")
list_time_lw = list()
for i in range(len(time_lw)):
    list_time_lw.append(time_lw.iloc[i]["temps_execution"])
tab_y_time = [(n+1)*1000 for n in range(999)]
pyplot.plot(tab_y_time, list_time_lw, color = 'red', linestyle = 'dashed', linewidth = 2)
pyplot.axis([-1000, 999999, 0, 10])
pyplot.title("Network Earthquake - Temps")
```
